# Moebius strip drawn by using SageMath

## Weierstrass data 

```python
g(z) = z^8*(z-1)/(z+1)
eta(z) = I*(z+1)^2/z^10
```

```python
phi1(z) = (1-g(z)^2)*eta(z)
phi2(z) = I*(1+g(z)^2)*eta(z)
phi3(z) = 2*g(z)*eta(z)
```

```python
Phi1(z) = integral(phi1(z),z)
Phi2(z) = integral(phi2(z),z)
Phi3(z) = integral(phi3(z),z)
```

```python
F(z) = ( Phi1(z).real(), Phi2(z).real(), Phi3(z).real() ) 
```

## Region 

```python
p(w) = exp(w) 
```

```python
var('y');
```

```python
eps = 0.00000001;
```

```python
xmin = -0.16;
xmax = 0;
ymin = eps;
ymax = pi-eps ;
nnx=30;
nny=110;
```

```python
nx = abs((xmin + xmax)/nnx)
ny = abs((ymin + ymax)/nny)
```

```python
points = [(p(x+I*y).real(), p(x+I*y).imag()) for x in srange(xmin,xmax,nx) for y in srange(ymin, ymax, ny, include_endpoint=True)] 
```

```python
list_plot(points, aspect_ratio=1)
```

```python
xlist = list( nx * k for k in range(0, nnx + 1) );
xlen= len(xlist);


ylist = list( ny * k for k in range(0, nny + 1) );
ylen=len(ylist);
```

## Plot the surface

```python
var('x y');
```

```python
X = lambda x: xlist[ int(round(x)) -1] if x<0.5 or x>xlen + 0.4 else xlist[ int(round(x)) -1]
Y = lambda y: ylist[ int(round(y)) -1] if y<0.5 or y>ylen + 0.4 else ylist[ int(round(y)) -1]
```

```python
f = (lambda x,y: Phi1(p(X(x)+ I*Y(y))).real(), 
     lambda x,y: Phi2(p(X(x)+ I*Y(y))).real(),
     lambda x,y: Phi3(p(X(x)+ I*Y(y))).real()
     )
gr1 = parametric_plot3d( f, (x, 1,xlen),(y,1,ylen),color="yellow")
```

```python
gr1.show(frame=True);
```

```python
gr2 = gr1.transform(rot=(0,1,0,pi));
```

```python
gr3 = gr1 + gr2 
```

```python
gr3.show(frame=False, aspect_ratio = (1,1,1))
```
